```python
import numpy as np
import math

def cost(x,y,w):
    sum1 = 0
    for i in range(len(x[0])):
        tmp = 0
        for j in range(len(x[i])):
            tmp += (w[1]*x[i][j] + w[0])
        sum1 += (tmp - y[i])**2
    return sum1/(2*len(x))
def func_grad(x,y,w,k):
    k+=1
    m=len(x)
    n=len(x[0])
    x=np.array(x)
    del_J=[]
    for j in range(n):
        out=[]
        for i in range(m):
            mul=0
            sub=0
            dot=0
            dot=np.dot(w,np.array(x[i]))
            sub=dot-y[i]
            mul=(sub*x[i][j]/m) if abs(sub*x[i][j]/m)<1e308 else (sub*x[i][j]/m)*1e308/abs(sub*x[i][j]/m) # edited this to avoid overflow warning
            if mul:
                out.append(mul)
        out=sum(out)
        del_J.append(out)
    return del_J

def grad_desc(x,y,w,e,alpha):
    k=0
    for i in range(len(x)):
        x[i].insert(0,1)
    #print('hi')
    while(True):
        k+=1
        del_J=np.array(func_grad(x,y,w,k))
        #print(del_J)
        if np.linalg.norm(np.array(del_J))<e or k==1000: # added a new constraint
            return w
        else:
            w=np.subtract(w,alpha*del_J)
```

```python
import numpy as np
import math

def cost_reg(x,y,w,reg):
    sum1 = 0
    for i in range(len(x[0])):
        tmp = 0
        for j in range(len(x[i])):
            tmp += (w[1]*x[i][j] + w[0])
        sum1 += (tmp - y[i])**2
    sum2=reg*sum([i**2 for i in w])
    return (sum1+sum2)/(2*len(x))
def func_grad_reg(x,y,w,k):
    k+=1
    m=len(x)
    n=len(x[0])
    x=np.array(x)
    del_J=[]
    for j in range(n):
        out=[]
        for i in range(m):
            mul=0
            sub=0
            dot=0
            dot=np.dot(w,np.array(x[i]))
            sub=dot-y[i]
            mul=(sub*x[i][j]/m) if abs(sub*x[i][j]/m)<1e308 else (sub*x[i][j]/m)*1e308/abs(sub*x[i][j]/m) # edited this to avoid overflow warning
            if mul:
                out.append(mul)
        out=sum(out)
        del_J.append(out)
    return del_J

def grad_desc_reg(x,y,w,e,alpha,reg):
    k=0
    m=len(x)
    #print('hi')
    while(True):
        k+=1
        del_J=np.array(func_grad_reg(x,y,w,k))
        #print(del_J)
        if np.linalg.norm(np.array(del_J))<e or k==5000: # added a new constraint
            return w
        else:
            w=np.subtract(w*(1-(alpha*reg/m)),alpha*del_J)
```

```python
import numpy as np

def PCA(x,n):
    C = x-np.mean(x,axis=0)
    cov_mat = np.cov(C, rowvar=False)
    e_values,e_vectors = np.linalg.eig(cov_mat)
    idx = np.argsort(e_values)[::-1]
    e_values = e_values[idx]
    e_vectors = e_vectors[:, idx]
    e_vectors = e_vectors[:, :n]
    res = np.dot(e_vectors.transpose(), C.transpose()).transpose()
    return res
    
    
X = np.array([[1,2,3],[2,3,4],[4,5,6],[3,4,5]])
num = 1
X_new = PCA(X, num)
print(X_new)
```
